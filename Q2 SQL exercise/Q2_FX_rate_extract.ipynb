Download FX rates for USD, GBP, CZK from Frankfurter API

```python
# Set the date range and currencies
start_date = "2024-03-01"
end_date = "2024-03-31"
base = "EUR"
symbols = "USD,GBP,CZK"

url = f"https://api.frankfurter.app/{start_date}..{end_date}?from={base}&to={symbols}"

# Get data
response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data['rates']).T
df.index.name = 'Date'
df = df.reset_index()
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# Create date range
date_range = pd.date_range(start="2024-03-01", end="2024-03-31", freq='D')

FX = pd.DataFrame({'Date': date_range})
FX = pd.merge(FX, df, on='Date', how='outer')
FX = FX.ffill()
FX
```

```python
FX['Date'] = FX['Date'].dt.strftime('%Y-%m-%d')
FX.to_csv('FX.csv')
```
